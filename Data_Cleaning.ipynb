#### Import Liabraries

```python
import pandas as pd
import numpy as np
import os
```

```python
pwd= os.getcwd()
```

```python
bellabeat_data= pd.read_csv(pwd+"\\daily_activity_merged.csv")
bellabeat_data.head()
```

```python
## Replace the "/" to "-" for consistency

bellabeat_data["ActivityDate"]= bellabeat_data["ActivityDate"].str.replace("/", "-")
bellabeat_data.head(2)
```

```python
## Check the shape of the data

bellabeat_data.shape
```

### <b> hourly_intensity_dataset <b>

```python
hourly_intensity_data= pd.read_csv(pwd+ "\\hourly_intensities_merged.csv")
hourly_intensity_data.head()
```

```python
## Replace the "/" to "-" for consistency

hourly_intensity_data["ActivityHour"] = hourly_intensity_data["ActivityHour"].str.replace("/", "-")
hourly_intensity_data.head(2)
```

```python
## Check the shape of the data

hourly_intensity_data.shape
```

```python
## Transform the datetime to only date

hourly_intensity_data["ActivityHour"]= pd.to_datetime(hourly_intensity_data["ActivityHour"]).dt.date
hourly_intensity_data.head(2)
```

```python
hourly_intensity_data.dtypes
```

```python
## Convert the "ActivityHour" column to datetime format

hourly_intensity_data["ActivityHour"]= pd.to_datetime(hourly_intensity_data["ActivityHour"])

## Create a new dataframe to extract the unique "Id" and "ActiveHours" and calculate the sum and mean

new_hourly_intensity_data= hourly_intensity_data.groupby(["Id", "ActivityHour"]).agg({
    "TotalIntensity": "sum",
    "AverageIntensity": "mean"
}).reset_index()

new_hourly_intensity_data.head()
```

```python
## Check the shape

new_hourly_intensity_data.shape
```

```python
## Concat this dataset with "bellabeat" dataset

bellabeat_data= pd.concat([bellabeat_data, new_hourly_intensity_data], axis=1)
bellabeat_data.head()
```

```python
## Drop the second "Id" column and "ActivityHour" column

bellabeat_data= bellabeat_data.drop(["Id", "ActivityHour"], axis=1)
bellabeat_data.head(2)
```

### <b> minute_METs_dataset <b>

```python
minute_METs_data= pd.read_csv(pwd+ "\\minute_METs_narrow_merged.csv")
minute_METs_data.head()
```

```python
## Change the "/" to "-" in "ActivityMinute" for consistency

minute_METs_data["ActivityMinute"]= minute_METs_data["ActivityMinute"].str.replace("/", "-")
minute_METs_data.head(2)
```

```python
## Change the "ActivityMinute" column from datetime to date

minute_METs_data["ActivityMinute"]= pd.to_datetime(minute_METs_data["ActivityMinute"]).dt.date
minute_METs_data.head(2)
```

```python
## Check the shape 

minute_METs_data.shape
```

```python
## Convert the ActivityMinute column to datetime format

minute_METs_data["ActivityMinute"]= pd.to_datetime(minute_METs_data["ActivityMinute"])

## Create a new dataframe to extract the unique "Id" and "ActivityMinute" and calculate the sum

new_minute_METs_data= minute_METs_data.groupby(["Id", "ActivityMinute"]).agg({
    "METs": "sum"
}).reset_index()

new_minute_METs_data.head()
```

```python
## Concat this dataset with "bellabeat" dataset

bellabeat_data= pd.concat([bellabeat_data, new_minute_METs_data], axis= 1)
bellabeat_data.head()
```

```python
## Drop the second "Id" and "ActivityMinutes" column

bellabeat_data= bellabeat_data.drop(["Id", "ActivityMinute"], axis= 1)
bellabeat_data.head(2)
```

### <b> sleep_day_merged dataset <b>

```python
sleep_day_data= pd.read_csv(pwd+ "\\sleep_day_merged.csv")
sleep_day_data.head()
```

```python
## Change the "/" to "-" in "ActivityMinute" for consistency

sleep_day_data["SleepDay"]= sleep_day_data["SleepDay"].str.replace("/", "-")
sleep_day_data.head(2)
```

```python
## Change the "ActivityMinute" column from datetime to date

sleep_day_data["SleepDay"]= pd.to_datetime(sleep_day_data["SleepDay"]).dt.date
sleep_day_data.head(2)
```

```python
## Convert the ActivityMinute column to datetime format

sleep_day_data["SleepDay"]= pd.to_datetime(sleep_day_data["SleepDay"])
sleep_day_data.dtypes
```

```python
## Check the shape 

sleep_day_data.shape
```

```python
## Concat this dataset with "bellabeat" dataset

bellabeat_data= pd.concat([bellabeat_data, sleep_day_data], axis= 1)
bellabeat_data.head()
```

```python
## Drop the second "Id" and "SleepDay" column

bellabeat_data= bellabeat_data.drop(["Id", "SleepDay"], axis= 1)
bellabeat_data.head()
```

```python
## check the columns 

bellabeat_data.columns
```

```python
## Check the datatypes

bellabeat_data.dtypes.to_frame()
```

```python
## Check if the dataset has null values

bellabeat_data.isna().sum()
```

```python
## Replace the NaN values with their means

TotalIntensity_mean= bellabeat_data["TotalIntensity"].mean()
bellabeat_data["TotalIntensity"] = bellabeat_data["TotalIntensity"].fillna(TotalIntensity_mean)

AverageIntensity_mean= bellabeat_data["AverageIntensity"].mean()
bellabeat_data["AverageIntensity"] = bellabeat_data["AverageIntensity"].fillna(AverageIntensity_mean)

METs_mean= bellabeat_data["METs"].mean()
bellabeat_data["METs"] = bellabeat_data["METs"].fillna(METs_mean)

TotalSleepRecords_mean= bellabeat_data["TotalSleepRecords"].mean()
bellabeat_data["TotalSleepRecords"] = bellabeat_data["TotalSleepRecords"].fillna(TotalSleepRecords_mean)

TotalMinutesAsleep_mean= bellabeat_data["TotalMinutesAsleep"].mean()
bellabeat_data["TotalMinutesAsleep"] = bellabeat_data["TotalMinutesAsleep"].fillna(TotalMinutesAsleep_mean)

TotalTimeInBed_mean= bellabeat_data["TotalTimeInBed"].mean()
bellabeat_data["TotalTimeInBed"] = bellabeat_data["TotalTimeInBed"].fillna(TotalTimeInBed_mean)
```

```python
## Check if the null values got replaced

bellabeat_data.isna().sum().to_frame()
```

```python
## Make ActivityDate column changed to datetime format

bellabeat_data["ActivityDate"]= pd.to_datetime(bellabeat_data["ActivityDate"])

## Make a column named "day_of_week"

bellabeat_data["weekday"]= bellabeat_data["ActivityDate"].dt.day_name()
bellabeat_data.head(2)
```

```python
## Performing basic statistical analysis

bellabeat_data.describe()
```

```python
## get some info about the dataset

bellabeat_data.info()
```

```python
## save the dataset

bellabeat_data.to_csv("cleaned_bellabeat_data.csv", index= False)
```
